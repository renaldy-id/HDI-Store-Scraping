```python
def getAndParseURL(url):
    page = req.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    return(soup)
```

```python
product_urls = [x.get("href") for x in soup.findAll("a", {"class" : "product-card__image-wrapper"})]
```

```python
product_urls
```

```python
name = []
cat = []
desc = []
img_url = []
# scrape data for every book URL: this may take some time
for url in product_urls:
    soup = getAndParseURL(url)
    # product name
    name.append(soup.find("div", class_ = re.compile("product-detail__name text-xl")).text)
    # product category
    cat.append(soup.find("span", class_ = re.compile("text-bold")).text)
    # product description
    desc.append(soup.find("div", class_ = "product-detail__description text-sm js-description-full").text)
    # image url
    img_url.append(soup.find("img", class_ = "product-detail__image").get("src"))
```

```python
prices = [x.text for x in soup.find_all("div", {"class" : "product-card__price text-base mb-1"})]
```

```python
prices
```

```python
import pandas as pd
```

```python
datascrap = pd.DataFrame({
    'Product Name' : name,
    'Price' : prices,
    'Category' : cat,
    'Description' : desc,
    'Image Url' : img_url})
```
